# GeoZones — Generate bounding boxes

Converts `zones_filtered.geojson` to `zones_bboxes.json`: a lightweight mapping of
zone ID → bounding box `[minx, miny, maxx, maxy]`.

```python
import geopandas as gpd
import shapely
import json, os

IN_FILE  = "./zones_filtered.geojson"
OUT_FILE = "./zones_bboxes.json"

zones = gpd.read_file(IN_FILE, engine="pyogrio")[["id", "geometry"]]
print(f"Loaded {len(zones):,} zones")

bounds = shapely.bounds(zones.geometry.values)  # (n, 4) array: minx, miny, maxx, maxy
bboxes = {
    zone_id: b.tolist()
    for zone_id, b in zip(zones["id"], bounds)
}

with open(OUT_FILE, "w") as f:
    json.dump(bboxes, f)

print(f"Done. {len(bboxes):,} bboxes → {OUT_FILE}  ({os.path.getsize(OUT_FILE)/1024/1024:.1f} MB)")
```
